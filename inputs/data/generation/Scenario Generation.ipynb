```python
import random

genders = ['m', 'f']
ages = ['baby', 10, 40, 70]
cost = ['rural', 'urban']
status = ['doctor', 'exec', 'farmer', 'artist', 'homeless']
altruism = ['donated a kidney','not on donor list']
fitness = ['pregnant','athlete','obese']
health = ['excellent health', 'average health','poor health']
sub_ab = ['drinker','smoker', 'drug addict','recovered']
law = ['criminal history', 'law abiding']
#disability = ['wheelchair-bound','blind',]

all_features = [genders, ages, cost, status, fitness, health, sub_ab, law, altruism]

for i in range(0, 61):
    k=random.randint(2,3)
    features_to_randomize = random.sample(all_features, k)
    random_features = []
    for f in features_to_randomize:
        random_features.append(random.choice(f))
    print(random_features)
```
